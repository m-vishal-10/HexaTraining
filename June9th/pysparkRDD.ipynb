```python

activity_counts = rdd.countByValue()

most_common_activity = max(
    activity_counts.items(),
    key=lambda item: item[1]
)

least_common_activity = min(
    activity_counts.items(),
    key=lambda item: item[1]
)

login_percentage = (
    rdd.filter(lambda activity: activity == "LOGIN").count()
    / rdd.count()
) * 100

video_play_percentage = (
    rdd.filter(lambda activity: activity == "VIDEO_PLAY").count()
    / rdd.count()
) * 100

print(f"Total Activities: {rdd.count()}")
print(f"Unique Activities: {rdd.distinct().count()}")
print(
    f"Most Common Activity  : {most_common_activity[0]} "
    f"({most_common_activity[1]} occurrences)"
)
print(
    f"Least Common Activity : {least_common_activity[0]} "
    f"({least_common_activity[1]} occurrences)"
)
print(f"LOGIN Activity %: {login_percentage:.2f}%")
print(f"VIDEO_PLAY Activity % : {video_play_percentage:.2f}%")
```
