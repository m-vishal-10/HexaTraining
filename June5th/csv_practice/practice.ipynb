```python
#Display employee names
import csv 

count = 0
with open("C:\\Users\\Asus\\OneDrive\\Desktop\\HexaTraining\\June5th\\csv_practice\\employees.csv","r") as f:
    reader = csv.reader(f)
    next(reader) # Skip header row
    for row in reader:
        count += 1
print(count)
```

```python
#highest salary 

highest_salary = 0
with open("C:\\Users\\Asus\\OneDrive\\Desktop\\HexaTraining\\June5th\\csv_practice\\employees.csv","r") as f:
    reader = csv.reader(f)
    next(reader) # Skip header row
    for row in reader:
        salary = int(row[3])  # Assuming salary is in the 4th column (index 3)
        if salary > highest_salary:
            highest_salary = salary
print(highest_salary)

  
```

```python
#find lowest salary
lowest_salary = float('inf')
with open("C:\\Users\\Asus\\OneDrive\\Desktop\\HexaTraining\\June5th\\csv_practice\\employees.csv","r") as f:
    reader = csv.reader(f)
    next(reader) 
    for row in reader:
        salary = int(row[3]) 
        if salary < lowest_salary:
            lowest_salary = salary
print(lowest_salary)
```

```python
#find avg salary 
avg_salary = 0
total_salary = 0
count = 0
with open("C:\\Users\\Asus\\OneDrive\\Desktop\\HexaTraining\\June5th\\csv_practice\\employees.csv","r") as f:
    reader = csv.reader(f)
    next(reader) 
    for row in reader:
        salary = int(row[3])  
        total_salary += salary
        count += 1
avg_salary = total_salary / count
print(avg_salary)
```
