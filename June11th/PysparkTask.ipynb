```python
patient_city_spending_df = appointments_df.groupBy("patient_id") \
    .agg(
        sum("consultation_fee").alias("total_fees")
    ) \
    .join(
        patients_df,
        on="patient_id"
    )

city_window = Window.partitionBy(
    "city"
).orderBy(
    col("total_fees").desc()
)

patient_city_spending_df.withColumn(
    "rank",
    rank().over(city_window)
).filter(
    col("rank") == 1
).show()
```

```python
patients_df = spark.read.csv(
    "patients.csv",
    header=True,
    inferSchema=True
)

appointments_df = spark.read.csv(
    "appointments.csv",
    header=True,
    inferSchema=True
)
```

```python
patients_df = patients_df.fillna({
    "city": "Unknown",
    "blood_group": "Not Available"
})

appointments_df = appointments_df.fillna({
    "consultation_fee": 0
})

preferences_df = preferences_df.select(
    "patient_id",
    "preferred_hospital",
    col("contact.phone").alias("phone"),
    col("contact.email").alias("email")
)

preferences_df = preferences_df.fillna({
    "phone": "Not Provided",
    "email": "Not Provided",
    "preferred_hospital": "Not Assigned"
})
```

```python
hospital_df = patients_df.join(
    appointments_df,
    on="patient_id",
    how="left"
).join(
    preferences_df,
    on="patient_id",
    how="left"
)

hospital_df.show()
```

```python
hospital_df = hospital_df.withColumn(
    "age_group",
    when(col("age") < 18, "Child")
    .when(col("age") < 40, "Adult")
    .otherwise("Senior")
)

hospital_df.show()
```

```python
hospital_df.select(
    "appointment_id",
    "department",
    "consultation_fee"
).show()

department_revenue_df = hospital_df.groupBy(
    "department"
).agg(
    sum("consultation_fee").alias("department_revenue")
)

department_revenue_df.show()
```

```python
patient_spending_df = hospital_df.groupBy(
    "patient_id",
    "patient_name"
).agg(
    sum("consultation_fee").alias("total_spending")
)

patient_spending_df.show()
```

```python
department_revenue_df = hospital_df.groupBy(
    "department"
).agg(
    count("appointment_id").alias("appointment_count"),
    sum("consultation_fee").alias("total_revenue"),
    avg("consultation_fee").alias("average_fee")
)

department_revenue_df.show()
```
