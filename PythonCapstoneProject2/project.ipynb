### Part 6  Functions

```python
def find_top_scorer(players):
    top = max(players, key=lambda p: int(p['runs']))
    return top['player_name'], int(top['runs'])

name, runs = find_top_scorer(players)
print(name, runs)
```

```python
def calculate_average_runs(players):
    return sum(int(p['runs']) for p in players) / len(players)

print(calculate_average_runs(players))
```

```python
def find_best_team(players):
    team_runs = {}
    for p in players:
        team_runs[p['team']] = team_runs.get(p['team'], 0) + int(p['runs'])
    best = max(team_runs, key=team_runs.get)
    return best, team_runs[best]

team, runs = find_best_team(players)
print(f"{team} : {runs} ")
```

```python
def find_total_boundaries(players):
    total_fours = sum(int(p['fours']) for p in players)
    total_sixes = sum(int(p['sixes']) for p in players)
    return total_fours, total_sixes

fours, sixes = find_total_boundaries(players)
print(f"total fours: {fours}")
print(f"total sixes: {sixes}")
print(f"total boundaries: {fours + sixes}")
```

### Part 7  exception handling

```python
def load_players(filename):
    players = []
    try:
        with open(filename, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                players.append(row)
        print("file loaded")
    except FileNotFoundError:
        print("error")
    return players
```

```python
def safe_get_runs(player):
    try:
        runs = int(player['runs'])
        if runs < 0:
            raise ValueError("runs cannot be negative")
        return runs
    except ValueError as e:
        print(e)
        return 0
```

```python
def safe_get_matches(player):
    try:
        matches = int(player['matches'])
        if matches <= 0:
            raise ValueError("match count cannot be negative")
        return matches
    except ValueError as e:
        print(e)
        return 0
```
