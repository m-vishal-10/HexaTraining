### Part 1 - File Handling

```python
#Task1 - read the csv file 
import csv
orders = []
try:
    with open('C:\\Users\\Asus\\OneDrive\\Desktop\\HexaTraining\\PythonCapstoneProject1\\orders.csv', 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            orders.append(row)
except FileNotFoundError:
    print("make sure the file path is correct and the file exists")
```

```python
#Task2- Display all the orders
for order in orders:
    print(order)
```

```python
#Task3 - Total Orders 
print(len(orders))
```

### Part 2 - Revenue Analysis

```python
#Task4 - Total Revenue
total_revenue = 0
for order in orders:
    total_revenue += int(order['quantity']) * int(order['price'])
print(total_revenue)
```

```python
#Task5 Highest Value Order
highest_value_order = 0 
order_id = 0
for order in orders:
    highest_value_order = max(highest_value_order, int(order['quantity']) * int(order['price']))
print(highest_value_order)
```

```python
#Task6 Lowest Value Order
lowest_value_order = float('inf')
order_id = 0
for order in orders:
    lowest_value_order = min(lowest_value_order, int(order['quantity']) * int(order['price']))
print(lowest_value_order)
```

```python
#Task7 Avg Value Order
total_value = 0
for order in orders:
    total_value += int(order['quantity']) * int(order['price'])
avg_value_order = total_value / len(orders)
print(round(avg_value_order, 2))
```

### Part 3 - Customer Analysis

```python
#Task8 Unique Customers
unique_cust = set()
for order in orders:
    unique_cust.add(order['customer_name'])
print(*unique_cust, sep="\n")
```

```python
#Task10 customer with highest value order
for order in orders:
    if int(order['quantity']) * int(order['price']) == highest_value_order:
        print(order)
```

### Part 4 - Product Analysis

```python
#task11 order of products
order_of_products = {}
for order in orders:
    product = order['product']
    order_of_products[product] = order_of_products.get(product, 0) + 1
order_of_products
```

```python
#task12 revenue of products
revenue_of_products = {}
for order in orders:
    product = order['product']
    price = int(order['price'])
    quantity = int(order['quantity'])
    revenue_of_products[product] = revenue_of_products.get(product, 0) + (price * quantity)
revenue_of_products
```

```python
#task13 most sold 
max_sold_value = max(order_of_products.values())
max_sold_product = [key for key, val in order_of_products.items() if val == max_sold_value]
print(*max_sold_product, sep='\n')
```

```python
#task14 least sold 
min_sold_value = min(order_of_products.values())
min_sold_products = [key for key, val in order_of_products.items() if val == min_sold_value]
print(*min_sold_products, sep='\n')
```

```python
#task 15 revenue by category 
revenue_by_category = {}
for order in orders:
    category = order['category']
    revenue_by_category[category] = revenue_by_category.get(category, 0) + (int(order['quantity']) * int(order['price']))
print(revenue_by_category)
```
