```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, sum, avg, max, min, count, when
```

```python
spark = SparkSession.builder.appName('DataframeExercises').getOrCreate()
```

```python
data = [
    (1, 'Rahul Sharma', 'Hyderabad', 'Electronics', 'Laptop', 1, 65000, 'Delivered'),
    (2, 'Priya Reddy', 'Bangalore', 'Electronics', 'Mobile', 2, 25000, 'Delivered'),
    (3, 'Amit Kumar', 'Mumbai', 'Furniture', 'Chair', 4, 3500, 'Pending'),
    (4, 'Sneha Patel', 'Chennai', 'Electronics', 'Tablet', 1, 30000, 'Cancelled'),
    (5, 'Farhan Ali', 'Delhi', 'Furniture', 'Table', 2, 12000, 'Delivered'),
    (6, 'Neha Singh', 'Hyderabad', 'Fashion', 'Shoes', 3, 2500, 'Delivered'),
    (7, 'Arjun Verma', 'Pune', 'Fashion', 'Watch', 1, 8000, 'Pending'),
    (8, 'Meera Nair', 'Bangalore', 'Electronics', 'TV', 1, 45000, 'Delivered'),
    (9, 'Kiran Rao', 'Hyderabad', 'Furniture', 'Desk', 1, 15000, 'Delivered'),
    (10, 'Ravi Kumar', 'Mumbai', 'Electronics', 'Mobile', 1, 28000, 'Delivered'),
]

columns = ['order_id', 'customer_name', 'city', 'category', 'product', 'quantity', 'price', 'order_status']

df = spark.createDataFrame(data, columns)
print('DataFrame created successfully!')
```

```python
df = df.withColumn(
    'order_value_category',
    when(col('total_amount') >= 50000, 'High')
    .when(col('total_amount') >= 20000, 'Medium')
    .otherwise('Low')
)
df.show()
```

```python
df.groupBy('order_value_category').count().orderBy('order_value_category').show()
```

```python
df.filter(col('order_status') == 'Delivered') \
  .select('order_id', 'customer_name', 'product', 'final_amount') \
  .show()
```

```python
df.filter(col('order_status') == 'Delivered') \
  .groupBy('city') \
  .agg(sum('final_amount').alias('delivered_revenue')) \
  .orderBy('city') \
  .show()
```
