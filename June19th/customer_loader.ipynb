```python
customer_data = [
    (1, "Vishal", "Chennai"),
    (2, "Rahul", "Bangalore"),
    (3, "Priya", "Chennai"),
    (4, "Anu", "Hyderabad"),
    (5, "Kiran", "Bangalore")
]

columns = ["customer_id", "customer_name", "city"]

customer_df = spark.createDataFrame(customer_data, columns)

customer_df.write.mode("overwrite").saveAsTable("customer_table")

print("Customer table created")
```
