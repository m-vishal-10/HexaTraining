```python
product_data = [
    (1, "Laptop", 60000),
    (2, "Mouse", 800),
    (3, "Keyboard", 1500),
    (4, "Monitor", 12000),
    (5, "Headphones", 2500)
]

columns = ["product_id", "product_name", "price"]

product_df = spark.createDataFrame(product_data, columns)

product_df.show()


```

```python
product_df.createOrReplaceTempView("products")
```

```python
%sql

SELECT AVG(price) AS average_price
FROM products
```
